# Title: Heart Failure Clinical Records Dataset

Source: The dataset was collected at the Faisalabad Institute of Cardiology and the Allied Hospital in
Faisalabad, Punjab, Pakistan, during the period of April to December in 2015. It was later made available on
Kaggle.


Description: The dataset comprises medical records of 299 patients diagnosed with heart failure. These records
include various clinical, body, and lifestyle information of the patients. The dataset encompasses 13 features,
providing insights into the patients' health status and potential risk factors for heart failure.

# Features:

Age: The age of the patient (numeric).

Anaemia: Indicates the presence of anemia (binary: 0 for no, 1 for yes).

Creatinine Phosphokinase (CPK): Level of creatinine phosphokinase enzyme in the blood (numeric).

Diabetes: Indicates whether the patient has diabetes (binary: 0 for no, 1 for yes).

Ejection Fraction: Percentage of blood leaving the heart at each contraction (numeric).

High Blood Pressure: Indicates whether the patient has high blood pressure (binary: 0 for no, 1 for yes).

Platelets: Platelet count in the blood (numeric).

Serum Creatinine: Level of creatinine in the blood (numeric).

Serum Sodium: Level of sodium in the blood (numeric).

Sex: Gender of the patient (binary: 0 for female, 1 for male).

Smoking: Indicates whether the patient smokes (binary: 0 for no, 1 for yes).

Time: Follow-up period (in days) for the patient's condition (numeric).

# 1. What is the distribution of age among heart failure patients in the dataset?

# 2. How does the death rate vary with age?

```python
# Check for missing values
print(df[['age','DEATH_EVENT']].isnull().sum())
```

```python
# Plot the death rates by age group
age_groups = df.groupby('age')['DEATH_EVENT'].mean().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(age_groups['age'], age_groups['DEATH_EVENT'], marker='o', linestyle='-', color='b')
plt.title('Death Rate by Age')
plt.xlabel('Age')
plt.ylabel('Death Rate')
plt.grid(True)
plt.show()
```

# 3. What is the percentage of male and female patients in the dataset?

```python
# Count the number of male and female patients
gender_counts = df['sex'].value_counts()

# Calculate percentages
total_patients = len(df)
male_percentage = (gender_counts[1] / total_patients) * 100  # assuming 1 = male
female_percentage = (gender_counts[0] / total_patients) * 100  # assuming 0 = female

# Display the results
print(f"Percentage of male patients: {male_percentage:.2f}%")
print(f"Percentage of female patients: {female_percentage:.2f}%")

# Data for plotting
percentages = [male_percentage, female_percentage]
genders = ['Male', 'Female']

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(genders, percentages, color=['blue', 'pink'])
plt.title('Percentage of Male and Female Patients')
plt.xlabel('Gender')
plt.ylabel('Percentage (%)')
plt.ylim(0, 100)  # Set the y-axis limit from 0 to 100
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 4.How does the platelet count vary among different age groups?

```python
age_bins = [0, 30, 40, 50, 60, 70, 80, 90]
age_labels = ['<30', '30-39', '40-49', '50-59', '60-69', '70-79', '80+']

# Bin the ages into groups
df['age'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

# Calculate average platelet count for each age group
platelet_means = df.groupby('age')['platelets'].mean()

# Plotting
plt.figure(figsize=(10, 6))
platelet_means.plot(kind='bar', color='skyblue')
plt.title('Average Platelet Count by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Platelet Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 5. Is there a correlation between creatine and sodium levels in the blood?

```python

# Calculate the correlation coefficient
correlation = df['serum_sodium'].corr(df['serum_creatinine'])
print(f"Correlation coefficient between CPK and Serum Sodium: {correlation:.2f}")

# Plotting
plt.figure(figsize=(10, 6))
sns.scatterplot(x='serum_sodium', y='serum_creatinine', data=df)
plt.title('Scatter Plot of Serum Creatinine vs. Serum Sodium')
plt.xlabel('Serum Sodium')
plt.ylabel('Creatinine Phosphokinase (CPK)')
plt.grid(True)
plt.show()
```

# 6.How does the prevalence of high blood pressure differ betweeen male and female patients?

```python
# Calculate prevalence of high blood pressure by gender
# Assuming 'High Blood Pressure' column is binary with 0 for no and 1 for yes
gender_pressure_counts = df.groupby('sex')['high_blood_pressure'].mean() * 100

# Mapping gender codes to labels
gender_labels = {0: 'Female', 1: 'Male'}
gender_pressure_counts.index = gender_pressure_counts.index.map(gender_labels)

# Plotting
plt.figure(figsize=(8, 6))
gender_pressure_counts.plot(kind='bar', color=['pink', 'blue'])
plt.title('Prevalence of High Blood Pressure by Gender')
plt.xlabel('Gender')
plt.ylabel('Percentage (%)')
plt.ylim(0, 100)  # Set y-axis limit from 0 to 100
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 7.What is the relationship between smoking habits and occurence of heart failure?

```python
# Calculate the prevalence of heart failure among smokers and non-smokers
# Assuming 'Smoking' column is binary with 0 for non-smokers and 1 for smokers
# and 'DEATH_EVENT' column is binary with 1 for occurrence of heart failure and 0 for non-occurrence
smoking_heart_failure = df.groupby('smoking')['DEATH_EVENT'].mean() * 100

# Mapping smoking codes to labels
smoking_labels = {0: 'Non-Smoker', 1: 'Smoker'}
smoking_heart_failure.index = smoking_heart_failure.index.map(smoking_labels)

# Plotting
plt.figure(figsize=(8, 6))
smoking_heart_failure.plot(kind='bar', color=['lightblue', 'salmon'])
plt.title('Prevalence of Heart Failure by Smoking Habit')
plt.xlabel('Smoking Habit')
plt.ylabel('Percentage of Heart Failure (%)')
plt.ylim(0, 100)  # Set y-axis limit from 0 to 100
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 8.Are there any noticeable patterns in the distribution of death events across different age groups?

```python

# Ensure that the Age column is numeric
df['age'] = pd.to_numeric(df['age'], errors='coerce')

# Define age bins and labels
age_bins = [0, 30, 40, 50, 60, 70, 80, 90]
age_labels = ['<30', '30-39', '40-49', '50-59', '60-69', '70-79', '80+']

# Bin the ages into groups
df['age'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

# Calculate the percentage of death events for each age group
death_event_distribution = df.groupby('age')['DEATH_EVENT'].mean() * 100

# Plotting
plt.figure(figsize=(10, 6))
death_event_distribution.plot(kind='bar', color='salmon')
plt.title('Distribution of Death Events Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Percentage of Death Events (%)')
plt.ylim(0, 100)  # Set y-axis limit from 0 to 100
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 9.Is there any significant in ejection fraction between patients with and without diabetes?

```python
from scipy import stats

# Load the dataset
data = pd.read_csv('heart_failure_clinical_records_dataset.csv')

# Group data by diabetes status
diabetes_groups = data.groupby('diabetes')['ejection_fraction']

# Extract ejection fractions for each group
ejection_fraction_diabetic = diabetes_groups.get_group(1)
ejection_fraction_non_diabetic = diabetes_groups.get_group(0)

# Perform an independent t-test to check for significant differences
t_stat, p_value = stats.ttest_ind(ejection_fraction_diabetic, ejection_fraction_non_diabetic, nan_policy='omit')

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Plotting
plt.figure(figsize=(10, 6))
sns.boxplot(x='diabetes', y='ejection_fraction', data=data, palette='Set2')
plt.title('Ejection Fraction by Diabetes Status')
plt.xlabel('Diabetes Status')
plt.ylabel('Ejection Fraction')
plt.xticks(ticks=[0, 1], labels=['Non-Diabetic', 'Diabetic'])
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

# 10.How does the serum creatinine level vary between patients who survived and those who did not?

```python
# Group data by survival status
survival_groups = data.groupby('DEATH_EVENT')['serum_creatinine']

# Extract serum creatinine levels for each group
serum_creatinine_survived = survival_groups.get_group(0)  # 0 for survived
serum_creatinine_died = survival_groups.get_group(1)      # 1 for did not survive

# Perform an independent t-test to check for significant differences
t_stat, p_value = stats.ttest_ind(serum_creatinine_survived, serum_creatinine_died, nan_policy='omit')

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Plotting
plt.figure(figsize=(10, 6))
sns.boxplot(x='DEATH_EVENT', y='serum_creatinine', data=data, palette='Set1')
plt.title('Serum Creatinine Levels by Survival Status')
plt.xlabel('Survival Status')
plt.ylabel('Serum Creatinine Level')
plt.xticks(ticks=[0, 1], labels=['Survived', 'Did Not Survive'])
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
